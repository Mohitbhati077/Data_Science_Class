```python
from tkinter import *
root=Tk()
root.title("Convert")
root.geometry("500x500+200+200")

def convert():
    try:
        dn=int(txtn1.get())
#     dn=int(dn)
    
        bn=bin(dn)
        oc=oct(dn)
        he=hex(dn)
        l2.config(text=f"Binary:{bn}")
        l3.config(text=f"Octal:{oc}")
        l4.config(text=f"Hexa:{he}")
    except ValueError:
         l2.config(text="Invalid input. Please enter a valid decimal number.")

    return

def cleardata():
    txtn1.delete(0,END)
    # l2.delete(0,END)
    # l3.delete(0,END)
    # l4.delete(0,END)
    txtn1.focus()
    return

l1=Label(root,text="Enter decimal number::",font="verdana 12")
l2=Label(root,text="Binary:")
l3=Label(root,text="Octal:")
l4=Label(root,text="Hexadecimal:")

txtn1=Entry(root,font="verdana 12")

btncon=Button(root,text="Convert",font="verdana",command=convert)
btnclear=Button(root,text="Clear",font="verdana",command=cleardata)
btnexit=Button(root,text="Exit",command=root.destroy)

l1.grid(row=0,column=0,padx=10,pady=10)
txtn1.grid(row=0,column=1,padx=10,pady=10)

l2.grid(row=1,column=0,padx=10,pady=10)

l3.grid(row=2,column=0,padx=10,pady=10)

l4.grid(row=3,column=0,padx=10,pady=10)

btncon.grid(row=4,column=0,padx=10,pady=10)
btnclear.grid(row=4,column=1,padx=10,pady=10)
btnexit.grid(row=4,column=2,padx=10,pady=10)

# def convert():
#     dn=entry.get()
#     dn=int(dn)
#     bn=bin(dn)
#     oc=oct(dn)
#     he=hex(dn)

root.mainloop()
    

```
